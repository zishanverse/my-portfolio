**1. The salaries of employees in an organization are normally distributed with mean 90000 and standard deviation 10000.**


```python
mean_salary = 90000    # mean salary
std_salary  = 10000    # standard deviation of salary
```

**1.a.** What is the probability of an employee having salary less than 75000?

```python
# Write your code here
# P(x < 75000)
prob_less_75000 = norm.cdf(75000, mean_salary, std_salary)
print(prob_less_75000)
```

**1.b.** What is the probability of an employee having salary more than 100000?

```python
# Write your code here
# P(x > 100000) = 1 - P(x <= 100000)
prob_more_100000 = 1 - norm.cdf(100000, mean_salary, std_salary)
print(prob_more_100000)
```

**1.c.** What is the probabilty of an employee salary between 80000 to 120000

```python
# Write your code here
# P(80000 <= x <= 120000) = P(x <= 120000) - P(x <= 80000)
prob_between_80000_120000 = norm.cdf(120000, mean_salary, std_salary) - norm.cdf(80000, mean_salary, std_salary)
print(prob_between_80000_120000)
```

**2. The scores(out of 1000) for an engineering entrance exam are normally distributed with a mean score of 500 and a standard deviation of 120.**

```python
mean_exam = 500      # mean of exma scores
std_exam  = 120      # Standard deviation of exam scores
```

**1.a.** What is the probability of a candidate scoring less than 600?

```python
# Write your code here
prob_less_600 = norm.cdf(600, mean_exam, std_exam)
print(prob_less_600)
```

**1.b.** What is the probability of a candidate scoring more than 700?

```python
# Write your code here
prob_more_700 = 1 - norm.cdf(700, mean_exam, std_exam)
print(prob_more_700)
```

**1.c.** What is the probability of a candidate scoring between 450 to 650?

```python
# Write your code here
prob_between_450_650 = norm.cdf(650, mean_exam, std_exam) - norm.cdf(450, mean_exam, std_exam)
print(prob_between_450_650)
```

**1.d.** Candidates falling in the bottom 35% will be considered as failing the test. How much a candidate has to score to pass?q

```python
# Write your code here
passing_score = norm.ppf(0.35, mean_exam, std_exam)

# Display result
print(passing_score)
```

**1.e.** The best engineering college only accepts top 0.1% of candidates. How much a candidate has to score to be in that college?


```python
# Write your code here
# let that score be x
# Probability of scoring above x = 0.001
top_college_score = norm.ppf(1 - 0.001, mean_exam, std_exam)
print(top_college_score)
```

**3. The mean age of population in a city is 35 with a standard deviation of 10. Assuming the age is normaly distributed answer the following questions.**

```python
mean_age = 35
std_age = 10
```

**1.a.** What percentage of residents have age below 50?

```python
# Write your code here
p_below_50 = norm.cdf(50, mean_age, std_age) * 100
print(p_below_50)
```

**1.b.** What percentage of residents have age above 30?

```python
# Write your code here
p_above_30 = (1 - norm.cdf(30, mean_age, std_age)) * 100
print(p_above_30)
```

**1.c.** What age corresponds to the second quartile?

```python
# Second quartile : The value below which 50 percent of residents are aged
# Or the age above which 50 percent of rersidents are aged

second_quartile = norm.ppf(0.50, mean_age, std_age)
print(second_quartile)
```

**1.d.** What age values make up the inter quartile range?


```python
# first quartile : the value below which 25% of residents are aged
# or the value above which 75% of residents are aged

first_quartile = norm.ppf(0.25, mean_age, std_age)
third_quartile = norm.ppf(0.75, mean_age, std_age)
interquartile_range = third_quartile - first_quartile

print(f"({first_quartile}, {third_quartile})")
```
