**1. Find the probability distribution (Probability of each outcome) for getting a five when rolling a dice four times.**

```
- Output Format:
Probability distribution for getting a five when rolling a dice 4 times:
P(X=0): ______
P(X=1): ______
P(X=2): ______
P(X=3): ______
P(X=4): ______
```

````python
# write your code here
n = 4       # number of dice rolls
p = 1/6     # probability of getting a five

print("Probability distribution for getting a five when rolling a dice 4 times:")
for k in range(n+1):
    prob = binom.pmf(k, n, p)
    print(f"P(X={k}): {round(prob, 4)}")
````

**2. A fair coin is tossed 10 times. Find out the probability of:**
1. Getting a head at most 6 times
2. Getting a head at least 3 times

```
- Output Format:
Coin Toss Probabilities:
Probability of getting at most 6 heads: ______
Probability of getting at least 3 heads: ______
```

````python
n = 10
p = 0.5

prob_at_most_6 = binom.cdf(6, n, p)
prob_at_least_3 = 1 - binom.cdf(2, n, p)

print("Coin Toss Probabilities:")
print(f"Probability of getting at most 6 heads: {round(prob_at_most_6, 4)}")
print(f"Probability of getting at least 3 heads: {round(prob_at_least_3, 4)}")
````

**3. A surgeon claims that the success rate of surgeries performed by him is 65%. What is the probability that out of next 100 patients:**
1. At least 60 patients will have a successful surgery
2. Fewer than 75 patients will have a successful surgery
3. There will be no more than 30 failed surgeries

```
- Output Format:
Surgery Success Rate Probabilities:
Probability of at least 60 successful surgeries: _____
Probability of less than 75 successful surgeries: _____
Probability of no more than 30 failed surgeries: _____
```

````python
n = 100
p = 0.65 # success rate
failure_p = 1 - p # failure rate

prob_at_least_60 = 1 - binom.cdf(59, n, p)
prob_less_than_75 = binom.cdf(74, n, p)
prob_no_more_30_failures = binom.cdf(30, n, failure_p) # Using failure rate and number of failures

print("Surgery Success Rate Probabilities:")
print(f"Probability of at least 60 successful surgeries: {prob_at_least_60:.4f}")
print(f"Probability of less than 75 successful surgeries: {prob_less_than_75:.4f}")
print(f"Probability of no more than 30 failed surgeries: {prob_no_more_30_failures:.4f}")
````

**5. A factory manufactures car batteries with a defect rate of 1.5%. In each shipment to a car company there are 10000 batteries. The cost paid for each battery by the the car company is 50000 rupees and the battery manufacturer does not provide any compensation for the defective batteries found in a shipment. Calculate the proability that the loss (in rupees) by the car company for a single shipment:**
1. Will not be more than 7500000
2. Will be between 6000000 to 9000000
3. Will be between 7000000 to 8000000

```
- Output Format:
Car Battery Defect Loss Probabilities:
Probability that loss will not be more than 7,500,000: ____
Probability that loss will be between 6,000,000 and 9,000,000: ____
Probability that loss will be between 7,000,000 and 8,000,000: ____
```

````python
import math
from scipy.stats import binom # Import binom

n = 10000
p = 0.015
q = 1 - p
cost = 50000

# Loss thresholds in terms of defective batteries
loss1 = 7500000
loss2_low = 6000000
loss2_high = 9000000
loss3_low = 7000000
loss3_high = 8000000

x1 = int(round(loss1 / cost)) # Round to nearest integer for Binomial
x2_low = int(round(loss2_low / cost))
x2_high = int(round(loss2_high / cost))
x3_low = int(round(loss3_low / cost))
x3_high = int(round(loss3_high / cost))

# Probabilities using Binomial distribution
prob1 = binom.cdf(x1, n, p)
prob2 = binom.cdf(x2_high, n, p) - binom.cdf(x2_low - 1, n, p) # Inclusive range
prob3 = binom.cdf(x3_high, n, p) - binom.cdf(x3_low - 1, n, p) # Inclusive range


print("Car Battery Defect Loss Probabilities:")
print(f"Probability that loss will not be more than 7,500,000: {prob1:.4f}")
print(f"Probability that loss will be between 6,000,000 and 9,000,000: {prob2:.4f}")
print(f"Probability that loss will be between 7,000,000 and 8,000,000: {prob3:.4f}")
````

**6. A factory produces `LED bulbs`, and `5% of them are defective`, A batch of 500 bulbs is inspected.**

**Find the probability that:**
- At most **20 bulbs** are defective.
- More than **30 bulbs** are defective.
- Between **10 to 25 bulbs** are defective.

```
- Output Format:
Defective Bulbs Probabilities:
Probability of at most 20 defective bulbs: _____
Probability of more than 30 defective bulbs: _____
Probability of defective bulbs between 10 to 25: _____
```

````python
mu = 500
sigma = 0.05

# Probabilities using normal approximation with continuity correction
prob_at_most_20 = binom.cdf(20.5, mu, sigma)
prob_more_than_30 = 1 - binom.cdf(30.5, mu, sigma)
prob_between_10_25 = binom.cdf(25.5, mu, sigma) - binom.cdf(9.5, mu, sigma)

print("Defective Bulbs Probabilities:")
print(f"Probability of at most 20 defective bulbs: {round(prob_at_most_20, 4)}")
print(f"Probability of more than 30 defective bulbs: {round(prob_more_than_30, 4)}")
print(f"Probability of defective bulbs between 10 to 25: {round(prob_between_10_25, 4)}")
````

**7. Email Spam Detection**
A spam filter correctly identifies **90% of spam emails**. Out of **100 emails**, assume that **40% are spam**.

  - **Find the probability that:**
      - At least **30 spam emails** are correctly identified.
      - At most **20 spam emails** are incorrectly classified as non-spam.

```
- Output Format:
Email Spam Detection Probabilities:
Probability of at least 30 spam emails correctly identified: _____
Probability of at most 20 spam emails incorrectly classified: _____
Probability of exactly 36 spam emails correctly identified: _____
```

````python
n = 40
p = 0.9

# At least 30 correctly identified
prob_at_least_30 = 1 - binom.cdf(29, n, p)

# At most 20 incorrectly classified → X ≥ 20
prob_at_most_20_incorrect = 1 - binom.cdf(19, n, p)

# Exactly 36 correctly identified
prob_exact_36 = binom.pmf(36, n, p)

print("Email Spam Detection Probabilities:")
print(f"Probability of at least 30 spam emails correctly identified: {prob_at_least_30:.4f}")
print(f"Probability of at most 20 spam emails incorrectly classified: {prob_at_most_20_incorrect:.4f}")
print(f"Probability of exactly 36 spam emails correctly identified: {prob_exact_36:.4f}")
````

**8. A machine has a 2% chance of failing each day. The factory has 50 such machines operating daily.**

- Find the probability that:
  - No more than 3 machines fail in a day.
  - More than 5 machines fail in a day.
  - Between 2 and 6 machines fail in a day.

```
- Output Format:
Machine Failure Probabilities:
Probability of no more than 3 failures: _____
Probability of more than 5 failures: _____
Probability of failures between 2 to 6: _____
```

````python
n = 50
p = 0.02

prob_no_more_3 = binom.cdf(3, n, p)
prob_more_5 = 1 - binom.cdf(5, n, p)
prob_between_2_6 = binom.cdf(6, n, p) - binom.cdf(1, n, p)

print("Machine Failure Probabilities:")
print(f"Probability of no more than 3 failures: {round(prob_no_more_3, 4)}")
print(f"Probability of more than 5 failures: {round(prob_more_5, 4)}")
print(f"Probability of failures between 2 to 6: {round(prob_between_2_6, 4)}")
````
