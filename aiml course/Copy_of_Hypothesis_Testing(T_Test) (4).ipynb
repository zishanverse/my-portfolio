**1. A nutritionist claims that the average calorie content in a brand of snack bars is 120 calories. To test this claim, a random sample of 10 snack bars is selected, and their calorie content is measured. The sample has a mean calorie content of 125 and a standard deviation of 10 calories. Conduct a one-sample t-test to determine if there is a significant difference between the sample mean and the claimed mean. Use a significance level of 0.05.** `Print the expected output in 3 decimals`

```python
pop_mean = 120
sample_size = 10
sample_mean = 125
sample_std = 10

#calculate Degrees of freedom
df = sample_size - 1

print('Sample Size : ', sample_size)
print('Sample Mean : ', sample_mean)
print('Sample STD  : ', sample_std)
print('Degrees of freedom : ', df)
```

```python
# Calculating t score (One sample test)
pop_mean = 120
sample_size = 10
sample_mean = 125
sample_std = 10

#calculate Degrees of freedom
df = sample_size - 1

# Calculate t score
t_value = (sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size))

# calculating critical t value
alpha = 0.05 / 2
t_critical = t.ppf(1 - alpha, df)

# Conclusion
if t_value > t_critical:
    result = "We have sufficient evidence to reject null hypothesis"
else:
    result = "We fail to reject null hypothesis"

print(f"T Value: {t_value:.3f}")
print(f"Critical T Value: {t_critical:.3f}")
print(f"Conclusion: {result}")
```

**2. A company wants to compare the productivity of two teams with unequal variances. Team A (n = 12) has a mean productivity score of 90 with a standard deviation of 5. Team B (n = 15) has a mean productivity score of 80 with a standard deviation of 6. Conduct an independent two-sample t-test without assuming equal variances at a significance level of 0.01. Use simplified degrees of freedom (df = n₁ + n₂ - 1).** `Print the expected output in 3 decimals`



```python
sample1_size = 12
sample1_mean = 90
sample1_std  = 5

sample2_size = 15
sample2_mean = 80
sample2_std = 6

df = sample1_size + sample2_size - 1

print('Sample1 Size : ', sample1_size)
print('Sample1 Mean : ', sample1_mean)
print('Sample1 STD  : ', sample1_std)
print('Sample2 Size : ', sample2_size)
print('Sample2 Mean : ', sample2_mean)
print('Sample2 STD  : ', sample2_std)
print('Degrees of freedom : ', df)
```

```python
# Calculating t score (Independent samples t test)
sample1_size = 12
sample1_mean = 90
sample1_std  = 5

sample2_size = 15
sample2_mean = 80
sample2_std = 6

df = sample1_size + sample2_size - 1

# calculate t value
t_value = (sample1_mean - sample2_mean) / np.sqrt((sample1_std**2 / sample1_size) + (sample2_std**2 / sample2_size))

# Critical t value
alpha = 0.01 / 2 # Two-tailed test
t_critical = t.ppf(1 - alpha, df)

# Conclusion
if abs(t_value) > t_critical:
    result = "We have sufficient evidence to reject null hypothesis."
else:
    result = "We fail to reject null hypothesis."

print(f"T Value: {t_value:.3f}")
print(f"Critical T Value: {t_critical:.3f}")
print(f"Conclusion: {result}")
```

**3. A researcher believes that a new drug has an effect on reducing cholesterol levels. The average cholesterol level in the population is 200 mg/dL. A sample of 15 patients who took the drug shows an average cholesterol level of 190 mg/dL with a standard deviation of 15 mg/dL. Conduct a one-sample t-test to determine if there is a significant reduction in cholesterol levels. Use a significance level of 0.01.** `Print the expected output in 3 decimals`

```python
pop_mean = 200
sample_size = 15
sample_mean = 190
sample_std = 15

#calculate Degrees of freedom
df = sample_size - 1

print('Sample Size : ', sample_size)
print('Sample Mean : ', sample_mean)
print('Sample STD  : ', sample_std)
print('Degrees of freedom : ', df)
```

```python
# Calculating t score (One sample test)
pop_mean = 200
sample_size = 15
sample_mean = 190
sample_std = 15

#calculate Degrees of freedom
df = sample_size - 1

# Calculate t score
t_value = (sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size))

# calculating critical t value for a one-tailed (left-tailed) test
alpha = 0.01
t_critical_for_display = t.ppf(1 - alpha, df) # Get positive critical value for display
t_critical_for_comparison = t.ppf(alpha, df) # Get negative critical value for comparison

# Conclusion for a left-tailed test
if t_value < t_critical_for_comparison:
    result = "We have sufficient evidence to reject null hypothesis."
else:
    result = "We fail to reject null hypothesis."

print(f"T Value: {abs(t_value):.3f}")
print(f"Critical T Value: {t_critical_for_display:.3f}")
print(f"Conclusion: {result}")
```

**4. A researcher wants to know if a new teaching method improves test scores. A class of 10 students takes a test before and after using the new method. The test scores before and after the method are as follows:**

**Before: 70, 65, 80, 75, 85, 90, 60, 55, 70, 75 <br>
After: 75, 68, 82, 78, 88, 95, 62, 58, 80, 68 <br>
Conduct a paired t-test to determine if there is a significant improvement in test scores. Use a significance level of 0.05.** `Print the expected output in 3 decimals`

```python
before = np.array([70, 65, 80, 75, 85, 90, 60, 55, 70, 75])
after  = np.array([75, 68, 82, 78, 88, 95, 62, 58, 80, 68])
```

```python
# calculate the t-score for a paired t-test
before = np.array([70, 65, 80, 75, 85, 90, 60, 55, 70, 75])
after  = np.array([75, 68, 82, 78, 88, 95, 62, 58, 80, 68])

differences = after - before
sample_size = len(before)

#calculate Degrees of freedom
df = sample_size - 1

mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=0)

t_value = mean_diff / (std_diff / np.sqrt(sample_size))

alpha = 0.05 / 2  # two tailed
t_critical = t.ppf(1 - alpha, df)



# Conclusion
if abs(t_value) > t_critical:
    result = "We have sufficient evidence to reject null hypothesis."
else:
    result = "We fail to reject null hypothesis."

print(f"T Value: {t_value:.3f}")
print(f"Critical T Value: {t_critical:.3f}")
print(f"Conclusion: {result}")
```

```python
sample = np.array([3.5, 3.8, 4.0, 3.9, 4.1,4.3])
pop_mean = 4
sample_size = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=0)

#calculate Degrees of freedom
df = sample_size - 1

print('Sample Size : ', sample_size)
print('Sample Mean : ', sample_mean)
print('Sample STD  : ', sample_std)
print('Degrees of freedom : ', df)
```

```python
# Calculating t score (One sample test)
sample = np.array([3.5, 3.8, 4.0, 3.9, 4.1,4.3])
pop_mean = 4
sample_size = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=0)

#calculate Degrees of freedom
df = sample_size - 1

# Calculate t score
t_value = abs((sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size)))

alpha = 0.05 / 2
t_critical = t.ppf(1 - alpha, df)

# Conclusion
if abs(t_value) > t_critical:
    result = "We have sufficient evidence to reject null hypothesis."
else:
    result = "We fail to reject null hypothesis."

print(f"T Value: {t_value:.3f}")
print(f"Critical T Value: {t_critical:.3f}")
print(f"Conclusion: {result}")
```
